# Curves to graph

- y = 0n +1  and label the curve "O(1)"
- y = log(n) and label the curve "O(log n)"
- y = n and label the curve "O(n)"
- y = n*log(n) and label it "O(n log n)"
- y = n^2 and label it "O(n^2)"
- y = 2^n and label it "O(2^n)"
- y = n! and label it "O(n!)"
- y = n^n and label it "O(n^n)"


```python
# import matplotlib.pyplot as plt
import math 
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

plt.suptitle("Big O Notation", c='green')

n = np.linspace(1, 20, 800)
plt.ylim(0, 100)



labels = ['O(1)', 'O(log n)', 'O(n)', 'O(n log n)', 'O(n^2)', 'O(2^n)', 'O(n^n)']
big_o = [np.ones(n.shape), np.log(n), n, n * np.log(n), n**2, 2**n, n**n]

for i in range(len(big_o)):
    plt.plot(n, big_o[i], label=labels[i])
    
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
```
