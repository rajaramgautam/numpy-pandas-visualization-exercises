# 1. Use matplotlib to plot the following equation:

# Add an anotation for the point 0, 0, the origin.

```python
import numpy as np
# generating data

x = range (-100,100)
y =[x**2 -x+2 for x in x]
# plot x & y
plt.plot(x, y, c='green')

#add label 
plt.xlabel("x axis")
plt.ylabel("y axis")
# add annotation 
plt.annotate('Origin', xy = (0,0), xytext = (0, 1000), arrowprops ={})

plt.show()
```

# 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

# You can use functions from the math module to help implement some of the equations above.


```python
x1 = range(0, 100)
y1 = [x**(1/2) for x in x1]
plt.plot(x1,y1)

plt.title(" $y = \sqrt{x}$")
plt.xlabel(" x axis")
plt.ylabel("y axis")
```

```python
x2 = range(0, 20)
y2 = [x**3 for x in x2]
plt.plot(x2,y2)
plt.title(" $y = x^3$")
plt.xlabel(" x axis")
plt.ylabel("y axis")
```

```python
x3 = range(1, 11)
y3 = [2**x for x in x3 ]
plt.plot(x3,y3)
plt.title(" $y = 2^x$")
plt.xlabel(" x axis")
plt.ylabel("y axis")
```

```python
x4 = range(1, 11)
y4 = [1/ (x+1) for x in x4]
plt.plot(x4,y4)
plt.title (" y = 1/(x+1)")
plt.xlabel(" x axis")
plt.ylabel("y axis")
```

# 3. Combine the figures you created in the last step into one large figure with 4 subplots.


```python
y1 = [x**(1/2) for x in x1]
y2 = [x**3 for x in x2]
y3= [x**3 for x in x3]
y4= y = [1/ (x+1) for x in x4]

plt.figure(figsize= (8,6))

plt.suptitle(" Combined Figures")

plt.subplot(2,2,1)

plt.plot(x1,y1, marker='_', c='green')
plt.grid(True, ls = '--')
plt.title(' y = x^(1/2)')

plt.subplot(2,2,2)
plt.plot(x2,y2, marker = '_', c='green')
plt.grid(True, ls = '--')
plt.title('y = x^3')

plt.subplot(2,2,3)
plt.plot(x3,y3, marker= '_', c='green')
plt.grid(True, ls = '--')
plt.title('y= 2^x')

plt.subplot(2,2,4)
plt.plot(x4,y4, marker= '*', c='green')
plt.grid(True, ls = '--')
plt.title('y= 1/(x+1)')
```
